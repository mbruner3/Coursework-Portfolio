### Defining the Problem & Discussing Factors

1. **Problem Definition**
    + **Problem to Solve:** Creating a reporting development schedule for FE <br>
    owned reports in-scope for the HCM project that will utilize all resources efficiently.
    + **Primary Objective:** *Minimize* the time for developing the FE own reports.

2. **Factors**  
    1. **Reporting Development Time:** Each resource can only spend so many hours <br> per week developing reports.
    2. **Module:** Resources have preferences for the module for developing reports.
    3. **Development Time:** Each report has a estimated time to develop that report<br> based upon the reports complexity.
    4. **Project Deadline:** All reports that are Priority 1 or 0 must be available<br> at January 1, 2023.
    5. **Report Experience:** Each reporting resource has different skill levels.

3. **Decision Variables**   
The decision variable represent the possibility of each of the 12 resources<br> to develop the 56 reports.

$$
x_{11},~x_{12},~x_{13},~x_{14},...,~x_{ij},~where~i={1,2,...,12}~and~j={1,2,...,64}
$$

4. **Objective Function**

$$
minimize:
\sum_{i=1}^{12}\sum_{j=1}^{64}c_{ij}x_{ij},~where~c_{ij}~and~x_{ij}\geq0
$$

5. **Module Constraints**  
The module's complexity is moderate, or complex.

$$
\begin{align}\\
    M_{1j} = \begin{cases}
      56,   &\text{ for j = 1, 2, 3}\\
      40, &\text{ for j = 4, 5, 6, ..., 22}\\
      0, &\text{ else}
    \end{cases} &
    M_{2j} = \begin{cases}
      40,   &\text{ for j = 23,24,...,27}\\
      0, &\text{ else}
    \end{cases} &
     M_{3j} = \begin{cases}
      40, &\text{ for j = 27, 28, 29, 30}\\
      0, &\text{ else}
    \end{cases} &
     M_{i4} = \begin{cases}
      56,   &\text{ for i = 5, 6}\\
      40, &\text{ for i = 31, 32, ...40}\\
      0, &\text{ else}
    \end{cases} &\\\\
     M_{i5} = \begin{cases}

    \end{cases} &
     M_{i6} = \begin{cases}

      40, &\text{ for i = 54, 55, ...56}\\
      0, &\text{ else}
    \end{cases}
\end{align}\\
$$

```python
# creating functions for scheduling reports based upon workdays and holidays.
def add_working_days(start_date, ADDED_DAYS):
    fe_holidays = [datetime(2022, 1, 1), datetime(2022, 1, 17), 
                   datetime(2022, 5, 30), datetime(2022, 7, 4),
                   datetime(2022, 9, 5), datetime(2022, 11, 11), 
                   datetime(2022, 11, 24),datetime(2022, 11, 25), 
                   datetime(2022, 12, 23), datetime(2022, 12, 26)]

    DAYS_ELAPSED = 0
    while DAYS_ELAPSED < ADDED_DAYS:
        test_date = start_date+timedelta(days=1)
        start_date = test_date
        if test_date.weekday()>4 or test_date in fe_holidays:
            # if a weekend or FE holiday, skip
            continue
        else:
            # if a workday, count as a day
            DAYS_ELAPSED += 1
    #creates the schedule for the build.
    return start_date

def subtract_working_days(start_date, SUBTRACTED_DAYS):
    fe_holidays = [datetime(2022, 1, 1), datetime(2022, 1, 17), 
                   datetime(2022, 5, 30), datetime(2022, 7, 4), 
                   datetime(2022, 9, 5), datetime(2022, 11, 11), 
                   datetime(2022, 11, 24), datetime(2022, 11, 25), 
                   datetime(2022, 12, 23), datetime(2022, 12, 26)]

    DAYS_ELAPSED = 0
    while DAYS_ELAPSED > SUBTRACTED_DAYS:
        test_date = start_date-timedelta(days=1)
        start_date = test_date
        if test_date.weekday()>4 or test_date in fe_holidays:
            # if a weekend or FE holiday, skip
            continue
        else:
            # if a workday, count as a day
            DAYS_ELAPSED += -1
    # creates the schedule for lean spec development
    return start_date
```

```python
# Variables for creating objective matrix.
x = [0]
y = [x] * NUM_MODULES

# Dictionary for storing the report complexities based upon modules.
coeff_list = {}

ITERATOR = 1
LEAD_ZEROS = 0

# Making a matrix based upon the number of modules and report complexities.
for i, module in enumerate(module_names):
    df = owners.get_group(module).value_counts('complexity')
    complex_names = list(df.index)
    report_numbers = list(df.values)
    z = [x] * LEAD_ZEROS
    y = [x] * (NUM_MODULES-ITERATOR)
    if i == 0: 
        for name, number in zip(complex_names, report_numbers):
            if name == 'Complex':
                j = 40
            elif name == 'Moderate':
                j = 20
            else:
                j = 10
            coeff_list[module.lower()+'_'+name.lower()] = np.repeat(z+[[j]]+y, number, axis = 1)

    elif i < (len(module_names)-1):
        ITERATOR += 1
        LEAD_ZEROS += 1
        z = [x] * LEAD_ZEROS
        y = [x] * (NUM_MODULES-ITERATOR)
        for name, number in zip(complex_names, report_numbers):
            if name == 'Complex':
                j = 40
            elif name == 'Moderate':
                j = 20
            else:
                j = 10
            coeff_list[module.lower()+'_'+name.lower()] = np.repeat(z+[[j]]+y, number, axis = 1)                 

    elif i == (len(module_names)-1):        
        ITERATOR = 0
        z = [x] * (len(module_names)-1)
        y = [x] * (ITERATOR)
        for name, number in zip(complex_names, report_numbers):
            if name == 'Complex':
                j = 40
            elif name == 'Moderate':
                j = 20
            else:
                j = 10
            coeff_list[module.lower()+'_'+name.lower()] = np.repeat(z+[[j]]+y, number, axis = 1)
```

```python
# Matrix for the objective function.
for i, coeff in enumerate(coeff_list):
    if i == 0:
        obj_matrix = coeff_list[coeff]
    else:
        obj_matrix = np.concatenate((obj_matrix, coeff_list[coeff]), axis = 1)

obj = [sum(x) for x in zip(*obj_matrix)]
obj = np.array(obj)
```

```python
# Cleaning module names.
mods = [re.sub('[\s]&[\s]|[\s]', '_', module.lower()) for module in module_names]

# This dictionary will store data relating to report complexity keys.
temp_list = list(coeff_list.keys())
complexity_keys = {}

for m in mods:
    module_complexity = []
    for i in temp_list:
        if m[:2]==i[:2]:
            module_complexity.append(i)
        complexity_keys[m]=module_complexity

columns = list(complexity_keys.values())
```

```python
# Resource matrices dictionary
resource_matrix = {}
for i in range(len(resource_info_df)):
    df = resource_info_df.loc[i,mods]
    resource_matrix[df.name] = np.hstack([
        coeff_list[j]*k for col, k in zip(columns,df) for j in col
])
    time_limits = list(resource_info_df.total_time_to_build_reports)

resource_names = list(resource_info_df.name.unique())
resource_ave_hours = list(resource_info_df[
    'how_many_hours_per_week_on_average_could_you_devote_to_report_development'
])
resource_start_dt = list(resource_info_df[
    'what_date_is_the_earliest_you_could_begin_building_reports'
])
standard_dev_time = list(resource_info_df['standard_reports_developed_days'])
moderate_dev_time = list(resource_info_df['moderate_reports_developed_days'])
complex_dev_time = list(resource_info_df['complex_reports_developed_days'])


report_names = list(fe_reports.ricew_object_name)
report_riceids = list(fe_reports.rice_id)
report_owners = list(fe_reports['functional_area_owners'])
report_complexity = list(fe_reports['complexity_for_oracle'])
build_start = list(fe_reports['technical_build_design_start_date'])
build_priority = list(fe_reports['report_priority_classification'])

REPORT_LENGTH = len(report_names)
RESOURCE_LENGTH = len(resource_names)

# Dictionary for storing resource matrices with experience included and time constraints.
resource_coeffs = {}
for j in range(RESOURCE_LENGTH):
    r = resource_matrix[j]
    r_array = np.array([sum(x) for x in zip(*r)])*resource_info['experience_weight'][j]
    resource_coeffs[j] = (r_array/resource_info['how_many_hours_per_week_on_average_could_you_devote_to_report_development'][j])*5

obj_coeffs = np.concatenate([resource_coeffs[j] 
                             for j in range(RESOURCE_LENGTH)], axis = 0)
```

```python
# Separating nonzero and zero coefficients from objective coeff matrix.
obj_coeffs_nonzero = obj_coeffs[obj_coeffs!=0]
obj_coeffs_zero = obj_coeffs[obj_coeffs==0]
```

```python
# Identifying all nonzero coeff in resource matrix.
resource_nonzero_coeffs = []
for i in resource_coeffs:
    nonzeros = resource_coeffs[i][resource_coeffs[i]!=0]
    resource_nonzero_coeffs.append(nonzeros)
```

```python
# Identifying all nonzero coeff in objective matrix.
reports_resources_coeffs = {}
for i in range(len(obj_coeffs)):
    if obj_coeffs[i]!=0:
        reports_resources_coeffs[resources_matrix[i]] = obj_coeffs[i]
```

```python
# Identifying all zero coeff in resource matrix.
resource_zero_coeffs = []
for i in resource_coeffs:
    zeros = resource_coeffs[i][resource_coeffs[i]==0]
    resource_zero_coeffs.append(zeros)
```

```python
# Identifying all zero coeff in objective matrix.
reports_resources_zero_coeffs = {}
for i in range(len(obj_coeffs)):
    if obj_coeffs[i]==0:
        reports_resources_zero_coeffs[resources_matrix[i]] = obj_coeffs[i]
```

```python
# Obtaining the keys for nonzero and zero resource matrices.
resources_nonzero_matrix = list(reports_resources_coeffs.keys())
resources_zero_matrix = list(reports_resources_zero_coeffs.keys())
```

```python
# Naming the resource nonzero and zero variables.
resource_vars = LpVariable.dicts('util', resources_nonzero_matrix, cat = 'Binary')
resource_zero_vars = LpVariable.dicts('util', resources_zero_matrix, cat = 'Binary')

# Creating the LP model.
model = LpProblem(name="report_scheduler", sense=LpMinimize)

# Objective Function.
model += (
    lpSum([reports_resources_coeffs[j]*resource_vars[j] for j in reports_resources_coeffs]),
    "Total Time to Develop Reports",
)

# Time limit constraint.
ITERATOR = 0
for i, resource in enumerate(resource_nonzero_coeffs):
    model += lpSum([obj_coeffs_nonzero[j]*resource_vars[resources_nonzero_matrix[j]] for j in range(ITERATOR, ITERATOR+len(resource))]) <= time_limits[i], resources[i]+'_time_limits'
    ITERATOR = ITERATOR+len(resource)

# Resource must have 2 reports constraint.
ITERATOR = 0
for i, resource in enumerate(resource_nonzero_coeffs):
    model += lpSum([resource_vars[resources_nonzero_matrix[j]] for j in range(ITERATOR, ITERATOR+len(resource))]) >= 2, resources[i]+'_report_lowlimits'
    ITERATOR = ITERATOR+len(resource)

report_list = []
for i in range(RESOURCE_LENGTH):
    report_list.append(i*REPORT_LENGTH)
report_list = np.array(report_list)

# Report can only be assigned to one and only one resource.
ITERATOR = 0
for report in reports:
    if ITERATOR >= REPORT_LENGTH:
        report_list_new = report_list+ITERATOR
        report_list_new = report_list_new[:3]
    else:
        report_list_new = report_list+ITERATOR
    report_list_index = []
    for i, j in enumerate(report_list_new):
        if obj_coeffs[j]==0:
            report_list_index.append(i)
    report_list_final = np.delete(report_list_new, report_list_index)
    model += lpSum([resource_vars[resources_matrix[j]] for j in report_list_final]) == 1, report+'_constraint'
    ITERATOR += 1

# I can only be assigned 38 reports.
ITERATOR = 0
for i, resource in enumerate(resource_nonzero_coeffs):
    if i != 8:
        ITERATOR = ITERATOR+len(resource)
    else:
        model += lpSum([resource_vars[resources_nonzero_matrix[j]] for j in range(ITERATOR, ITERATOR+len(resource))]) <= 38, resources[i]+'_constraint'
        ITERATOR = ITERATOR+len(resource)

# Resources cannot be assigned reports that are not in a preference module.
ITERATOR = 0
for i, resource in enumerate(resource_zero_coeffs): 
        model += lpSum([resource_zero_vars[resources_zero_matrix[j]] for j in range(ITERATOR, ITERATOR+len(resource))]) == 0, resources[i]+'_zero_constraint'
        ITERATOR = ITERATOR+len(resource)
print(model.solve())

print("Status:", LpStatus[model.status])

# The optimised objective function value is printed to the screen
print("Total Time for Reports = ", value(model.objective))
```

```python
complexity_dummies = pd.get_dummies(report_schedule_df['report_complexity'])
report_schedule_df = pd.concat([report_schedule_df, complexity_dummies], axis = 1)
report_schedule_length = len(report_schedule_df)

complexity = ['Standard/Basic', 'Moderate', 'Complex']
complexity_dummies_cols = list(complexity_dummies.columns)
for i in complexity:
    if i not in complexity_dummies_cols:
        missing = pd.DataFrame([[0]]*report_schedule_length, columns = [i])
        complexity_dummies = pd.concat([complexity_dummies, missing], axis = 1, names = i)

dev_times = []

report_dev_times = report_schedule_df[['complex_dev_time', 'moderate_dev_time', 'standard_dev_time']]
for i in range(len(report_schedule_df)):
    dev_times.append(complexity_dummies.iloc[i][0]*report_dev_times.iloc[i][0]+complexity_dummies.iloc[i][1]*report_dev_times.iloc[i][1]+complexity_dummies.iloc[i][2]*report_dev_times.iloc[i][2])

dev_times_df = pd.DataFrame(dev_times, columns = ['report_develop_days'], index = report_schedule_df.index)    
report_schedule_df= pd.concat([report_schedule_df, dev_times_df], axis = 1)

report_schedule_group = report_schedule_df.groupby('report_resource')

final_schedule = pd.DataFrame()
scheduled_resources = list(report_schedule_df.report_resource.unique())

for name in scheduled_resources:
    df = report_schedule_group.get_group(name)
    df = df.sort_values('build_priority')
    dev_times = df['report_develop_days']
    deadlines = []
    build_start_date = []
    lean_spec_end = []
    lean_spec_start = []
    for i, d in enumerate(dev_times):
        if i == 0:
            date = max(df['build_start'])
            deadlines.append(add_working_days(date, d))
            build_start_date.append(date)
            lean_spec_end.append(subtract_working_days(date, -1))
            lean_spec_start.append(subtract_working_days(date, -6))
        else:
            date = deadlines[i-1]
            days_out = timedelta(days=int(d))
            deadlines.append(add_working_days(date, d))
            build_start_date.append(date)
            lean_spec_end.append(subtract_working_days(date, -1))
            lean_spec_start.append(subtract_working_days(date, -6))

    deadlines_df = pd.DataFrame(deadlines, columns = ['build_end_date'], index = df.index)
    build_start_date_df = pd.DataFrame(build_start_date, columns = ['build_start_date'], index = df.index)
    lean_spec_end_df = pd.DataFrame(lean_spec_end, columns = ['lean_spec_end_date'], index = df.index)
    lean_spec_start_df = pd.DataFrame(lean_spec_start, columns = ['lean_spec_start_date'], index = df.index)

    report_schedule_final = pd.concat([df, lean_spec_start_df, lean_spec_end_df, build_start_date_df, deadlines_df], axis = 1)
    final_schedule = pd.concat([final_schedule, report_schedule_final], axis = 0)

    final_schedule_all_rpts = final_schedule.sort_values(['report_resource', 'build_start'])
```
